```python
import requests
import json
from bs4 import BeautifulSoup
import csv
```

```python
import re
from itertools import zip_longest
result_name = []
result_url = []
result_address = []
for i in range(15):
    dynamic_url = "https://sr-search.com/syarousi-search/result/?pg=" + str(i+1) + "&area%5B0%5D=11&area%5B1%5D=12&area%5B2%5D=13&area%5B3%5D=14"
    #print(dynamic_url)
    response = requests.get(dynamic_url)
    text = response.text
    soup = BeautifulSoup(text, "html.parser")
    for a in soup.select('.w500 a '):
        #print(a)
        ###会社名###
        title = re.findall('title=".*"', str(a))
        string = ",".join(title)
        title_2 = string.replace('title="', '')
        title_3 = title_2.replace('"', '')
        print(title_3)
        result_name.append(title_3)
    
    
        ###URL情報###
        personal_url = a['href']
        #print(personal_url)
        url = 'https://sr-search.com' + str(personal_url)
        print(url)
        print("----------------------------")
        result_url.append(url)
        

        ###住所###
        
    for i in soup.select('.list01 '):
        ###住所###
        address = re.findall('地</th>\s<td>.*</td>', str(i))
        #print(address)
        string = ",".join(address)
        address_2 = string.replace('地</th>\n<td>', '')
        address_3 = address_2.replace('</td>', '')
        print("----------------------------")
        print(address_3)
        result_address.append(address_3)

        
print("----------------------------")
#print(result_name)
#print(len(result_name))
print("----------------------------")
#print(result_url)
#print(len(result_url))


result_final= [result_name, result_url, result_address]
export_data = zip_longest(*result_final, fillvalue = '')
####csv ファイル書き出し###################
for item in result_name:
    with open('C:\\Scraping\\sharoshi_list.csv', 'a', newline="") as f:
        writer = csv.writer(f)
        print(item)
        writer.writerows(export_data)
```
